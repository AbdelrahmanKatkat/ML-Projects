### 1 Overview of Decision Tree

**Decision Tree Defination**:

Decision Trees (DTs) are a non-parametric supervised learning method used for classification and regression. The goal is to create a model that predicts the value of a target variable by learning simple decision rules inferred from the data features. A tree can be seen as a piecewise constant approximation.


**1. Decision Tree:**

  A Decision Tree is a versatile supervised machine learning algorithm used for both **classification and regression tasks**.

  It models data as a **tree-like structure** where each **internal node represents a decision based on a particular feature**, and each **leaf node represents the predicted outcome**.

  Here's a breakdown:

  1. **Introduction:**

      - Decision Trees work by recursively **splitting the dataset into subsets based on the values of different features**.
      
      - The process continues until a stopping criterion is met, such as a maximum depth or a certain number of data points in a node.

  2. **Nodes and Edges:**
    - **Root Node:** The topmost node that represents the entire dataset.
    - **Internal Nodes:** Nodes representing decisions based on features.
    - **Leaf Nodes:** Terminal nodes that predict the outcome or class.

  3. **Splitting Criteria:**
    - Decision Trees use various criteria to decide how to split data at each internal node, such as Gini impurity, entropy, or mean squared error.
    - The goal is to reduce uncertainty and improve the purity of classes in resulting subsets.

  4. **Building Process:**
    - Decision Trees are constructed through **a top-down, greedy approach**.
    - At each internal node, the algorithm chooses the best feature and threshold to split the data, aiming to **maximize the information gain or impurity reduction**.

  5. **Stopping Conditions:**
    - Trees can overfit, capturing noise in the data. Stopping conditions like maximum depth or minimum samples per leaf help prevent this.

  6. **Advantages:**
    - Easy to understand and visualize.
    - Can handle numerical and categorical data.
    - Suitable for feature selection.
    - Non-parametric model (doesn't assume data follows a specific distribution).

  7. **Disadvantages:**
    - Prone to **overfitting**, especially with deep trees.
    - Can be **unstable due to small changes in data leading to different tree structures**.

#### 1.4 Theory Behind Decision Tree

**Theory behind Decision Trees:**

A Decision Tree is a non-linear, supervised machine learning algorithm that makes decisions by partitioning the data into subsets based on the values of input features. The algorithm aims to create a tree-like structure that can be used for classification or regression tasks.

**Key Concepts:**

1. **Entropy and Information Gain:**

  Information gain is a measure of the reduction in entropy that results from splitting a dataset based on a particular feature. Entropy is a measure of the randomness or uncertainty in a dataset, and it is calculated based on the frequency of different classes or outcomes in the data.
  
  When we split a dataset based on a feature, we **create subsets of the data that are more homogeneous with respect to the target variable** (i.e., the variable we are trying to predict). This means that the resulting subsets have lower entropy than the original dataset, since there is less uncertainty about the likely outcome for each subset.

  Information gain **measures the difference between the entropy of the original dataset and the weighted average of the entropies of the subsets created by the feature split**. If the feature split results in subsets that are more homogeneous than the original dataset, then the information gain will be high, indicating that the feature is a good candidate for splitting. On the other hand, if the feature split does not result in much reduction in entropy, then the information gain will be low, indicating that the feature is less useful for splitting.

  Thus, information gain is closely related to entropy, as it quantifies the degree to which a feature split reduces entropy and increases our knowledge about the target variable.

   - **Entropy:** Entropy is a measure of impurity or uncertainty in a dataset. For a binary classification problem, the entropy formula is:

     $$\text{Entropy}(S) = -p_1 \log_2(p_1) - p_2 \log_2(p_2)$$

     where \(p_1\) and \(p_2\) are the proportions of two classes in dataset \(S\).

   - **Information Gain:** At each node of the tree, the algorithm calculates the information gain by subtracting the weighted average of child node entropies from the parent node entropy. A higher information gain suggests a better split.  information gain is used to determine the importance of each feature in the dataset, based on how much it contributes to the overall reduction in entropy across all the decision trees in the forest. The features with higher information gain are considered more important, as they have a greater influence on the final predictions made by the model.

2. **Gini Impurity:**

  Gini impurity is a measure of the degree of impurity or randomness in a dataset. It is commonly used in decision tree algorithms to evaluate the quality of a split at a given node in the tree.

  The **Gini impurity of a dataset is calculated as the probability of misclassifying a randomly chosen element in the dataset** if it were labeled randomly according to the class distribution in the dataset. In other words, if we randomly assign a label to an element in the dataset based on the proportion of each class in the dataset, then the Gini impurity is the probability that this label will be incorrect.

  A **Gini impurity value of 0 indicates that all elements in the dataset belong to the same class, while a value of 1 indicates that the elements are evenly distributed among all classes**. A lower Gini impurity value indicates that the dataset is more homogeneous with respect to the target variable, and therefore easier to classify accurately.

  In decision tree algorithms, the Gini impurity is used to evaluate the quality of a split at a given node by calculating the weighted average of the Gini impurities of the child nodes resulting from the split. A split that results in child nodes with lower Gini impurity values is considered to be a better split, as it leads to more homogeneous subsets of data and better classification accuracy.

   - Gini impurity measures the probability of a randomly chosen element being misclassified. For a binary classification problem, the Gini impurity formula is:

     $$\text{Gini}(S) = 1 - p_1^2 - p_2^2$$

   - Like entropy, Decision Trees aim to minimize Gini impurity when choosing split points.

3. **Splitting Criterion:**

   - Decision Trees use entropy, Gini impurity, or other criteria to decide the best feature and threshold for splitting data at each node.
   - The algorithm calculates the information gain or reduction in impurity for each potential split and chooses the one with the highest gain.

4. **Stopping Criteria:**

   - To prevent overfitting, Decision Trees have stopping criteria. Common criteria include reaching a maximum depth, having a minimum number of samples in a node, or achieving a certain purity threshold.

**Mathematics behind Decision Trees:**

The mathematics of Decision Trees involves calculating information gain or impurity reduction to find the optimal feature and threshold for splitting the data. Let's break down the process mathematically:

1. **Entropy Calculation:**

   Given a set \(S\) with \(p_1\) samples of class 1 and \(p_2\) samples of class 2, the entropy is calculated as:

   $$\text{Entropy}(S) = -p_1 \log_2(p_1) - p_2 \log_2(p_2)$$

2. **Information Gain:**

   Information Gain (\(IG\)) for a split using feature \(A\) is calculated as the difference between the entropy of the parent node and the weighted average of entropies of child nodes:

   $$IG(A) = \text{Entropy}(S) - \sum_{v \in \text{values}(A)} \frac{|S_v|}{|S|} \cdot \text{Entropy}(S_v)$$

3. **Gini Impurity Calculation:**

   For Gini impurity (\(G\)) calculation, similar to entropy, we calculate the impurity for the parent node and the impurity reduction due to a split:

   $$\text{Gini}(S) = 1 - p_1^2 - p_2^2$$

   $$\text{Gini Gain}(A) = \text{Gini}(S) - \sum_{v \in \text{values}(A)} \frac{|S_v|}{|S|} \cdot \text{Gini}(S_v)$$

4. **Choosing Best Split:**

   The feature and threshold that maximize information gain or reduce Gini impurity the most are selected as the optimal split points.

These mathematical calculations are performed recursively at each node during the construction of the Decision Tree, resulting in a tree structure that makes decisions based on these criteria.

In summary, Decision Trees leverage mathematical formulas for entropy, information gain, and Gini impurity to make optimal decisions when splitting data, leading to the construction of a tree-like model for classification or regression tasks.

### 2 Overview of Random Forest

**Defination**

An ensemble learning method for classification, regression and other tasks that operates by constructing a multitude of decision trees at training time. For classification tasks, the output of the random forest is the class selected by most trees. For regression tasks, the mean or average prediction of the individual trees is returned

**2. Random Forest:**

  A Random Forest is an **ensemble learning method that combines multiple Decision Trees** to improve predictive accuracy and control overfitting. It's a widely used algorithm with several key characteristics:

  1. **Introduction:**

      Random Forest builds multiple Decision Trees on various subsets of the training data, using both:

    1. **Random Feature Selection:**
      - At each split in a Decision Tree, **a random subset of features is considered** for the best split.
      - This **reduces correlation between trees** and decorrelates their predictions.

    2. **Bagging:**
      - Bagging involves **training each tree on a bootstrapped subset of the training data** (sampling with replacement).
      - Combining predictions from multiple trees helps **reduce overfitting and increase generalization**.

  4. **Ensemble Prediction:**
    - For **classification**, each tree votes, and the class with the **most votes** is the ensemble prediction.
    - For **regression**, the ensemble prediction is the **average of individual** tree predictions.

  5. **Advantages:**
    - Improved predictive accuracy compared to individual Decision Trees.
    - Handles high-dimensional data well.
    - Reduces overfitting and provides more stable predictions.

  6. **Disadvantages:**
    - Can be computationally expensive due to multiple trees.
    - Interpretability decreases as the number of trees increases.
    - Requires tuning hyperparameters like the number of trees and maximum depth.

#### 2.3 Theory Behind Random Forest

**Theory behind Random Forest:**

Random Forest is an ensemble learning technique that combines multiple Decision Trees to create a more robust and accurate predictive model. It reduces overfitting and increases generalization by aggregating predictions from different trees.

**Key Concepts:**

Random feature selection refers to the process of selecting a random subset of features from the dataset at each node in the decision tree. This helps to reduce overfitting and increase the diversity of the trees, as different subsets of features may be selected for different trees.

Bagging, on the other hand, involves creating multiple decision trees using bootstrap samples of the original dataset. Each tree is trained on a different subset of the data, which helps to reduce the variance and improve the accuracy of the model.

In random forest, we combine these two techniques by creating a set of decision trees, each of which is trained on a bootstrap sample of the data and a random subset of features. By using both random feature selection and bagging together, we can create a diverse set of trees that capture a wider range of patterns and relationships in the data, while also reducing overfitting and improving the accuracy of the model.

1. **Bootstrap Aggregating (Bagging):**

   - Random Forest uses bagging, a technique that involves creating multiple subsets of the training data by sampling with replacement. Each subset is used to train a separate Decision Tree.

2. **Random Feature Selection:**

   - At each split in a Decision Tree, only a subset of features is considered for the split.
   - This introduces randomness and ensures that trees are decorrelated, reducing the risk of overfitting.

3. **Ensemble Prediction:**

   - For classification, each Decision Tree in the Random Forest votes for the predicted class, and the class with the most votes is the ensemble prediction.
   - For regression, the ensemble prediction is the average of the predictions from individual trees.

4. **Out-of-Bag (OOB) Samples:**

   - Since bagging involves random sampling with replacement, some samples are left out in each bootstrap iteration. These out-of-bag samples can be used to estimate model performance without the need for a separate validation set.

**Mathematics behind Random Forest:**

The mathematics behind Random Forest involves the principles of bagging and random feature selection. Here's a breakdown of the mathematical aspects:

1. **Bagging:**

   - Given a dataset \(D\) with \(N\) samples, multiple bootstrap samples \(D_1, D_2, ..., D_B\) are created, each containing \(N\) samples randomly selected with replacement from \(D\).

2. **Random Feature Selection:**

   - At each node split in a Decision Tree, only a subset of features (\(m\)) is considered.
   - The parameter \(m\) can be specified or tuned, controlling the number of features to be considered.

3. **Ensemble Prediction:**

   - For classification, each Decision Tree's predicted class contributes to the ensemble prediction. The class with the most votes becomes the final prediction.
   - For regression, the ensemble prediction is the average of the predictions from all the trees.

4. **Out-of-Bag (OOB) Error:**

   - OOB samples, which were not included in the training set for a specific tree, can be used to estimate the model's generalization performance.
   - The OOB error is the average error over all trees on their respective OOB samples.

In summary, Random Forest combines bagging and random feature selection to create an ensemble of Decision Trees. The ensemble prediction improves accuracy and robustness by aggregating predictions from multiple trees. The mathematics involve concepts of sampling, averaging, and aggregation to create a more powerful predictive model.

## 4 Parameters of DT & RF

#### 4.1 DT Parameters

**Parameters for Decision Trees:**

1. **Criterion:**
   - This parameter defines the criterion for choosing splits. Common values are "gini" (Gini impurity) and "entropy" (information gain).
   - For Gini impurity (G), the formula is: \( G = 1 - p_1^2 - p_2^2 \), where \( p_1 \) and \( p_2 \) are the proportions of two classes in the node.

2. **Max Depth:**
   - Limits the maximum depth of the tree. Useful to control overfitting.
   - Deeper trees can fit training data better but might overfit.

3. **Min Samples Split:**
   - Minimum number of samples required to split a node further.
   - Prevents further splitting when the number of samples is below this threshold.

4. **Min Samples Leaf:**
   - Minimum number of samples required in a leaf node.
   - Prevents creation of tiny leaves with very few samples.

5. **Max Features:**
   - Controls the number of features considered for a split.
   - It can be a specific number or a proportion of the total features.



**Parameters for Decision Tree:**

1. **max_depth:**
   - Controls the maximum depth of the tree.
   - Setting it too high can lead to overfitting.

2. **min_samples_split:**
   - Minimum number of samples required to split an internal node.
   - Increasing it can lead to simpler trees.

3. **min_samples_leaf:**
   - Minimum number of samples required to be at a leaf node.
   - Prevents creation of small, uninformative leaves.

#### 4.2 RF Parameters

**Parameters for Random Forests:**

1. **N_estimators:**
   - Number of Decision Trees in the ensemble.
   - Increasing trees can lead to better accuracy, up to a certain point.

2. **Criterion:**
   - Similar to Decision Trees, defines the criterion for choosing splits.
   - "gini" or "entropy" are common choices.

3. **Max Depth:**
   - Limit the maximum depth of individual trees in the ensemble.

4. **Min Samples Split:**
   - Minimum number of samples required to split a node in each tree.

5. **Min Samples Leaf:**
   - Minimum number of samples required in a leaf node in each tree.

6. **Max Features:**
   - Controls the number of features considered for a split in each tree.
   - In random forest, max_features is a hyperparameter that controls the maximum number of features that can be used for splitting at each node in the decision tree. By default, max_features is set to the square root of the total number of features in the dataset, but it can be set to a specific value or a fraction of the total number of features.

  - Setting max_features to a smaller value can help to reduce overfitting and increase the diversity of the trees, as different subsets of features may be selected for different trees. However, setting it too low may also lead to underfitting and reduced accuracy, as important features may be excluded from consideration.

7. **Bootstrap:**
   - Controls whether bootstrap samples (sampling with replacement) are used to train trees.
   - Set to True for typical Random Forest behavior.



**Parameters for Random Forest:**

1. **n_estimators:**
   - Number of trees in the Random Forest.
   - Increasing it can improve performance, but there's a point of diminishing returns.

2. **max_depth:**
   - Maximum depth of individual trees in the ensemble.

3. **min_samples_split:**
   - Minimum number of samples required to split a node in each tree.

4. **min_samples_leaf:**
   - Minimum number of samples required in a leaf node in each tree.

5. **max_features:**
   - Number of features to consider when looking for the best split at each node.

6. **bootstrap:**
   - Controls whether bootstrap samples are used for training.

```python
from sklearn.ensemble import RandomForestClassifier

# Creating a Random Forest with specified parameters
rf_classifier = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    min_samples_split=5,
    min_samples_leaf=2,
    max_features='sqrt',
    bootstrap=True
)

# Fit the model
rf_classifier.fit(X_train, y_train)

# Predictions
rf_predictions = rf_classifier.predict(X_test)
```

## 5 Grid Search

Performing grid search for both Random Forest and Decision Tree involves systematically exploring different combinations of hyperparameters to find the best combination that yields the highest performance on a validation set. Here's how you can perform grid search for each algorithm and select the right parameters:

**Grid Search for Decision Tree:**

1. **Define Parameter Grid:**
   - Create a grid of hyperparameters you want to search. For a Decision Tree, this might include "criterion," "max_depth," "min_samples_split," and "min_samples_leaf."

2. **Create Cross-Validation Split:**
   - Split your training data into training and validation sets. Consider using k-fold cross-validation to get a reliable estimate of performance.

3. **Grid Search:**
   - For each combination of hyperparameters in the grid, train a Decision Tree on the training set and evaluate its performance on the validation set.

4. **Select Best Parameters:**
   - Choose the hyperparameters that resulted in the best performance on the validation set.

**Selecting Right Parameters for Decision Tree:**

- **Criterion:**
  - Choose based on your dataset. Gini impurity is more robust to noisy data, while entropy tends to create more balanced trees.

- **Max Depth:**
  - Start with a wide range and narrow down based on grid search results. Avoid too deep trees to prevent overfitting.

- **Min Samples Split and Min Samples Leaf:**
  - Smaller values make the tree more flexible but increase the risk of overfitting. Tune based on performance.

**Grid Search for Random Forest:**

1. **Define Parameter Grid:**
   - Similar to Decision Trees, create a grid of hyperparameters including "n_estimators," "criterion," "max_depth," "min_samples_split," and "min_samples_leaf."

2. **Create Cross-Validation Split:**
   - Split your training data into training and validation sets using k-fold cross-validation.

3. **Grid Search:**
   - For each combination of hyperparameters, train a Random Forest on the training set and evaluate its performance on the validation set.

4. **Select Best Parameters:**
   - Choose the hyperparameters that yielded the best performance on the validation set.

**Selecting Right Parameters for Random Forest:**

- **N_estimators:**
  - Higher values generally lead to better generalization, but with diminishing returns. Start with a reasonable range.

- **Criterion:**
  - Similar to Decision Trees, choose based on your dataset's characteristics.

- **Max Depth, Min Samples Split, Min Samples Leaf:**
  - Similar to Decision Trees, tune these parameters to control the complexity and overfitting of individual trees.

**Selecting Right Parameters in General:**

- **Validation Performance:**
  - Use a separate validation set or cross-validation to measure performance for different parameter combinations.

- **Avoid Overfitting:**
  - Parameters like max_depth, min_samples_split, and min_samples_leaf help control overfitting. Be cautious not to over-tune.

- **Trade-offs:**
  - There's often a trade-off between bias and variance. Simpler models might underfit, while complex models might overfit.

- **Domain Knowledge:**
  - Consider domain knowledge and data characteristics when selecting parameters.

In summary, grid search involves systematically exploring hyperparameters to find the best combination for optimal model performance. Choose parameters that balance model complexity, overfitting, and performance on validation data.

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Create a Decision Tree classifier
dt_classifier = DecisionTreeClassifier()

# Perform grid search with cross-validation
grid_search = GridSearchCV(dt_classifier, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

# Get the best parameters and the best estimator
best_params = grid_search.best_params_
best_dt_classifier = grid_search.best_estimator_

# Evaluate on the validation set
accuracy = best_dt_classifier.score(X_val, y_val)
print("Best Parameters:", best_params)
print("Validation Accuracy:", accuracy)
````
